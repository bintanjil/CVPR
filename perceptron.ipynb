```python
import matplotlib.pyplot as plt
import random
import math

x1=[0,0,1,1]
x2=[0,1,0,1]
y=[0,0,0,1]

plt.figure()

for i in range(len(y)):
    if y[i]==0:
        plt.plot(x1[i],x2[i],"ro")
    else:
        plt.plot(x1[i],x2[i],"go")
plt.show()
        
```

```python
w=[random.random(),random.random()]
b=random.random()
print(w,b)

theta=0.5
lr=0.001

# def sigmoid(x):
#     return 1 / (1 + math.exp(-x))
```

```python
epoch=0
count=0

#print(f"Epoch : {epoch}")
while count< 4:
    total_error=0
    print(f"Epoch : {epoch}")
    for i in range(len(x1)):
        weighted_sum=w[0]*x1[i]+w[1]*x2[i]+b
        pred=0
        if weighted_sum>theta:
            pred+=1
        error=y[i]-pred
        if error!=0:
            count=0
            total_error+=math.sqrt(error**2)
            w[0] +=lr*error*x1[i]
            w[1] +=lr*error*x1[i]
            b+=lr*error
        else :
            count+=1
       
        print(f"[{x1[i]} , {x2[i]}] = {pred}")
    print("===------------------------------===")
    print(f"Loss : {total_error}")
    epoch+=1
    



```
